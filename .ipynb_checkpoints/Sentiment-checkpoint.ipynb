We will be using TextBlob to measure sentiment and polarity

```python
from textblob import TextBlob
import pandas as pd

data = pd.read_pickle('pickle\\data_clean.pkl')
data
```

```python
# We will make two lambda functions for polarity and subjectivity
    # .apply() method applies a lambda function for each row in a column
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

data['polarity'] = data['transcripts'].apply(pol)
data['subjectivity'] = data['transcripts'].apply(sub)
data
```

```python
for movie in (data.index):
    print(data.loc[movie].polarity)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 8]

for movie in (data.index):
    x = data.loc[movie].polarity
    y = data.loc[movie].subjectivity
    plt.scatter(x, y, color='blue')
    plt.text(x+.001, y+.001, movie, fontsize=10)
    plt.xlim(.04, .18)
    
plt.title('Sentiment Analysis', fontsize=20)
plt.xlabel('<-- Negative -------- Positive -->', fontsize=15)
plt.ylabel('<-- Facts -------- Opinions -->', fontsize=15)
    
plt.show()
```

```python
import numpy as np
import math

def split_text(text, n=20):
    # We first determine the length of the text
        # Let's say 1000 in this case
    length = len(text)
    # Now, we'll determine what size chunks we want to grab from the text
        # If n = 10, each chunk is 100
    size = math.floor(length / n)
    # We now create an array that consists of chunks in steps the size of 'size'
        # Thus, until it reaches length (1000), the array is:
            #[0, 100, 200, 300...1000]
    start = np.arange(0, length, size)
    
    # Pull out equally sized pieces of text and put into a list
    split_list = []
    for piece in range(n):
        split_list.append(text[start[piece]:start[piece]+size])
        # So at piece [0], it takes the text chunk of 0 - 100 char
            # This is bc start[0] = 0 and start[1] = 100
        # Then moves to next piece, which is:
            # start[1] = 100 and start[2] = 200
    return split_list
```

```python
piece_dict = {}
pol_dict = {}
for movie in data.index:
    piece_dict[movie] = split_text(data.loc[movie].transcripts)
    
for key, value in piece_dict.items():
    newkey = key
    newvalue = [TextBlob(i).sentiment.polarity for i in value]
    pol_dict[key] = newvalue
    
pol_dict
```

```python
plt.rcParams['figure.figsize'] = [16, 12]

for index, movie in enumerate(data.index):    
    plt.subplot(6, 4, index+1)
    plt.plot(pol_dict[movie])
    plt.plot(np.arange(0,20), np.zeros(20))
    plt.title(movie)
    plt.ylim(ymin=-.3, ymax=.45)
    
plt.show()
```
