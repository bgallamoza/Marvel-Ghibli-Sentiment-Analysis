```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy

%matplotlib inline
```

```python
data = pd.read_pickle('pickle\\data_sent.pkl')
data.index.name = 'movie'
data
```

```python
data_melt = pd.melt(data, #dataframe
                   ['transcripts', 'producer'], #Variables to KEEP
                   var_name='test',
                   ignore_index=False) #Every other var is melted to this

data_melt
```

```python
sns.set_style(style='whitegrid')
g = sns.FacetGrid(data=data_melt,col='test',sharey=False,col_order=['polarity','subjectivity'])
g.map(sns.barplot, x='producer',
                y='value',
                data=data_melt,
     order=data_melt.producer.unique(),
     hue_order=data_melt.producer.unique,
     palette=sns.color_palette(['red','blue']),
     capsize=.3,
     errwidth=1.2,
     alpha=.7)
g.set_axis_labels('Producer','Value')
g.set_titles('average {col_name} of movies')
```
